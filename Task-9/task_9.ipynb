```python
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].mean())
df['sqft_living'] = df['sqft_living'].fillna(df['sqft_living'].mode()[0])
df['sqft_lot'] = df['sqft_lot'].fillna(df['sqft_lot'].mean())
df['floors'] = df['floors'].fillna(df['floors'].mean())
df['waterfront'] = df['waterfront'].fillna(df['waterfront'].mean())
df['view'] = df['view'].fillna(df['view'].mean())
df['condition'] = df['condition'].fillna(df['condition'].mean())
df['sqft_above'] = df['sqft_above'].fillna(df['sqft_above'].mean())
df['sqft_basement'] = df['sqft_basement'].fillna(df['sqft_basement'].mean())
df['yr_built'] = df['yr_built'].fillna(df['yr_built'].mean())
df['yr_renovated'] = df['yr_renovated'].fillna(df['yr_renovated'].mean())
df['street']=df['street'].fillna(df['street'].mode()[0])
df.isnull().sum().sort_values(ascending=False)
```
